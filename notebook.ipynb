## Exploratory Data Analysis (EDA)

```python
# Membuat kolom ["timestamp"] untuk menampung kolom ["year", "month", "day", "hour"]
file_paths = ["Aotizhongxin Station.csv", "Dongsi Station.csv", "Tiantan Station.csv"]
three_stations = []

for path in file_paths:
    df = pd.read_csv(path)
    df['timestamp'] = pd.to_datetime(df[['year', 'month', 'day', 'hour']])
    three_stations.append(df)

# gabungkan semua dataframe berdasarkan timestamp
merged_data = pd.concat(three_stations, axis=0, ignore_index=True)
merged_data.info()
```

```python
# Membuat fungsi yang melakukan kalkulasi AQI berdasarkan standar AQI Tiongkok dari
# konsentrasi polutan PM2.5 dan PM10
def calculate_aqi_china_pm25_pm10(concentration):
    if concentration <= 35:
        return (50/35) * concentration
    elif concentration <= 75:
        return ((50/(75 - 35)) * (concentration - 35)) + 50
    elif concentration <= 115:
        return ((50/(115 - 75)) * (concentration - 75)) + 100
    elif concentration <= 150:
        return ((50/(150 - 115)) * (concentration - 115)) + 150
    elif concentration <= 250:
        return ((50/(250 - 150)) * (concentration - 150)) + 200
    elif concentration <= 350:
        return ((50/(350 - 250)) * (concentration - 250)) + 300
    elif concentration <= 500:
        return ((50/(500 - 350)) * (concentration - 350)) + 400
    else:
        return 500

# Membuat dan menggabungkan kolom baru untuk menampung AQI dari konsentrasi polutan PM2.5 dan PM10
merged_data["AQI_PM2.5"] = merged_data["PM2.5"].apply(calculate_aqi_china_pm25_pm10)
merged_data["AQI_PM10"] = merged_data["PM10"].apply(calculate_aqi_china_pm25_pm10)
```

```python
# Melihat jenis berbagai kolom dan tipe datanya, terlihat ada kolom unnamed: 0 yang tidak dibutuhkan,
# kolom tersebut muncul ketika 3 file station di merge.
merged_data.info()
```

```python
# Menghapus kolom ["Unnamed: 0"] karena tidak ada kegunaan
merged_data.drop("Unnamed: 0", axis=1)
```

## Visualization & Explanatory Analysis

```python
# Dari AQI yang telah dihitung tadi, pada bagian Explanatory dan Visualization akan ditampilkan
# bentuk chartnya, dibagi menjadi 2 chart yaitu untuk AQI PM2.5 dan AQI PM10
plt.figure(figsize=(12,7))
sns.boxplot(x="station", y="AQI_PM2.5", data=merged_data)
plt.title("Boxplot AQI PM2.5 di Setiap Stasiun")
plt.xlabel("Stasiun")
plt.ylabel("AQI PM2.5")
plt.show()
```

```python
# Berikut merupakan chart kedua yaitu visualisasi AQI PM10 di ketiga stasiun
plt.figure(figsize=(12,7))
sns.boxplot(x="station", y="AQI_PM10", data=merged_data)
plt.title("Boxplot AQI PM10 di Setiap Stasiun")
plt.xlabel("Stasiun")
plt.ylabel("AQI PM10")
plt.show()
```

### Pertanyaan 2:

```python
# Membuat fungsi untuk mengklasifikasikan kualitas udara jika konsentrasinya bernilai x,
# terdapat 2 fungsi karena PM2.5 dan PM10 memiliki standar batas bawah yang berbeda.
def classification_aqi_pm25(concentration):
    if concentration <= 35:
        return "Kualitas Udara Sangat Baik"
    elif concentration <= 75:
        return "Kualitas Udara Baik"
    elif concentration <= 115 :
        return "Kualitas Udara Tercemar(Ringan)"
    elif concentration <=150 :
        return "Kualitas Udara Tercemar(Sedang)"
    elif concentration <= 250:
        return "Kualitas Udara Tercemar(Berat)"
    else :
        return "Kualitas Udara Sangat Tercemar"

def classification_aqi_pm10(concentration):
    if concentration <= 50:
        return "Kualitas Udara Sangat Baik"
    elif concentration <= 150:
        return "Kualitas Udara Baik"
    elif concentration <= 250 :
        return "Kualitas Udara Tercemar(Ringan)"
    elif concentration <= 350 :
        return "Kualitas Udara Tercemar(Sedang)"
    elif concentration <= 420 :
        return "Kualitas Udara Tercemar(Berat)"
    else :
        return "Kualitas Udara Sangat Tercemar"
```

```python
# Dari fungsi yang telah dibuat diatas, diaplikasikan ke kolom ["PM2.5"] dan ["PM10"], lalu ditampung dalam
# sebuah variabel baru yang memiliki nama ["Status_AQI_<concentration>"].
merged_data["Status_AQI_PM2.5"] = merged_data["PM2.5"].apply(classification_aqi_pm25)
merged_data["Status_AQI_PM10"] = merged_data["PM10"].apply(classification_aqi_pm10)
```

```python
# Setelah itu, melakukan groupby berdasarkan statusnya, dan menampilkan jumlah dari masing-masing
# kategori
classification_counts_pm25 = merged_data.groupby("Status_AQI_PM2.5").size()
classification_counts_pm10 = merged_data.groupby("Status_AQI_PM10").size()
print(classification_counts_pm25)
print(classification_counts_pm10)
```

```python
# Kemudian, dilakukan reset_index() untuk memastikan indeks yang telah dilakukan groupby kembali menjadi
# DataFrame seperti semula.
classification_counts_pm25_df = classification_counts_pm25.reset_index(name="counts")
classification_counts_pm10_df = classification_counts_pm10.reset_index(name="counts")
```

```python
# Melakukan visualisasi data untuk menjawab pertanyaan ke-2,
fig, ax = plt.subplots()
ax.bar(classification_counts_pm25.index, classification_counts_pm25.values)
plt.xticks(rotation=45)
plt.ylabel("Count pada ketiga stasiun")
plt.xlabel("Status AQI PM2.5")
plt.show()
```

```python
# Melakukan visualisasi data untuk menjawab pertanyaan ke-2,
fig, ax = plt.subplots()
ax.bar(classification_counts_pm10.index, classification_counts_pm10.values)
plt.xticks(rotation=45)
plt.ylabel("Count pada ketiga stasiun")
plt.xlabel("Status AQI PM10")
plt.show()
```
